```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.action_chains import ActionChains
import time

# Path to your ChromeDriver
driver_path = "path/to/chromedriver"

# Initialize the WebDriver
service = Service(driver_path)
driver = webdriver.Chrome(service=service)

try:
    # Open the webpage
    driver.get("URL_OF_YOUR_WEBPAGE")
    
    # Wait for the page to load
    wait = WebDriverWait(driver, 10)
    
    # Fill out the first name
    first_name_field = wait.until(EC.presence_of_element_located((By.NAME, "first_name")))
    first_name_field.send_keys("Your First Name")
    
    # Fill out the last name
    last_name_field = driver.find_element(By.NAME, "last_name")
    last_name_field.send_keys("Your Last Name")
    
    # Fill out the address
    address_field = driver.find_element(By.NAME, "address")
    address_field.send_keys("Your Address")
    
    # Fill out the phone number
    phone_number_field = driver.find_element(By.NAME, "phone_number")
    phone_number_field.send_keys("Your Phone Number")
    
    # Fill out the pin code
    pin_code_field = driver.find_element(By.NAME, "pin_code")
    pin_code_field.send_keys("Your Pin Code")
    
    # Select state from dropdown (example for a dropdown using its visible text)
    state_dropdown = driver.find_element(By.NAME, "state")
    state_dropdown.click()
    time.sleep(1)  # Wait for dropdown options to appear
    state_option = driver.find_element(By.XPATH, "//option[@value='Your State']")
    state_option.click()
    
    # Check the GST option (assuming radio buttons)
    gst_option = driver.find_element(By.ID, "gst_yes")  # Use the actual ID or selector for GST 'Yes'
    gst_option.click()
    
    # Check all the required terms and conditions checkboxes
    checkboxes = driver.find_elements(By.CSS_SELECTOR, "input[type='checkbox']")
    for checkbox in checkboxes:
        if not checkbox.is_selected():
            checkbox.click()
    
    # Click the Pay Now button
    pay_now_button = driver.find_element(By.XPATH, "//button[contains(text(), 'Pay now')]")
    pay_now_button.click()
    
    print("Form submitted successfully!")

except Exception as e:
    print(f"An error occurred: {e}")

finally:
    # Close the browser
    time.sleep(5)
    # driver.quit()
```
