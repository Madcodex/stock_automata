```python
import pandas as pd 
import os
from datetime import datetime, timezone
import hashlib
import numpy as np
```

```python
def get_ses_id(vid_name):


    if type(vid_name)==type(np.nan):
        return vid_name
    
    data = {}
    vid_name = vid_name.strip()

    data["headers"] = {}
    data["headers"]["device_id"] = vid_name.split("_")[1]
    data["headers"]["source"] = "cfc1"
    date_time = vid_name.split("_")[2] + vid_name.split("_")[3]
    # print(vid_name)
    try:

        if("." in vid_name):
            if ('+' in vid_name) or ('-' in vid_name):
                data["headers"]["start_time"] = datetime.strptime(date_time, "%Y%m%d%H%M%S.%f%z").astimezone(timezone.utc).strftime("%Y-%m-%d %H:%M:%S.%f")
            else:
                data["headers"]["start_time"] = datetime.strptime(date_time, "%Y%m%d%H%M%S.%f").astimezone(timezone.utc).strftime("%Y-%m-%d %H:%M:%S.%f")
            
        else:
            data["headers"]["start_time"] = datetime.strptime(date_time, "%Y%m%d%H%M%S%z").astimezone(timezone.utc).strftime("%Y-%m-%d %H:%M:%S.%f")
        data["headers"]["session_id"] = hashlib.sha256((data["headers"]["device_id"] + data["headers"]["start_time"]).encode("UTF-8")).hexdigest()
    except Exception as e:
        print(vid_name)
        raise e

    return data["headers"]["session_id"]
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
import time
from tqdm import tqdm
from concurrent.futures import ThreadPoolExecutor
from threading import Lock 
# Setup
chrome_driver_path = r"C:\Users\akash.verma\Downloads\chromedriver-win64\chromedriver-win64\chromedriver.exe"
chrome_options = Options()
chrome_options.debugger_address = "127.0.0.1:8892"  # Reuse an existing browser instance
service = Service(chrome_driver_path)

# Initialize a single WebDriver instance
driver = webdriver.Chrome(service=service, options=chrome_options)
```

```python

lock = Lock()  # To synchronize access to the single browser instance

# Function to download video for a single event ID
def download_video(event_id):
    with lock:  # Ensure only one thread interacts with the browser at a time
        try:
            # Open a new tab
            driver.execute_script("window.open('');")
            driver.switch_to.window(driver.window_handles[-1])  # Switch to the new tab

            # Load the event URL
            url = f"https://pui-rde-petivity.com/cat-feeder/events/{event_id}"
            driver.get(url)
            time.sleep(10)  # Allow time for the page to fully load

            try:
                # Find and click the download button
                download_button = driver.find_element(By.XPATH, "//a[contains(text(), 'Download Video')]")
                download_button.click()
                print(f"Download initiated for event ID {event_id}")
                time.sleep(1)  # Wait for the download to start
            except Exception as e:
                print(f"Could not find download button for event ID {event_id}: {e}")
            finally:
                # Close the tab after completing the task
                # driver.close()
                driver.switch_to.window(driver.window_handles[0])  # Switch back to the first tab
        except Exception as e:
            print(f"Error occurred for event ID {event_id}: {e}")


with ThreadPoolExecutor(max_workers=1) as executor:  # Adjust max_workers based on system capacity
    list(tqdm(executor.map(download_video, event_ids[:]), total=len(event_ids[:])))
```
