```python
from PIL import Image, ImageEnhance, ImageFilter
import pytesseract

def preprocess_image(image_path):
    # Open the image
    image = Image.open(image_path)
    
    # Convert to grayscale
    image = image.convert("L")
    # cv2.imshow('image',image)
    # cv2.waitKey(0)
    # Increase contrast
    enhancer = ImageEnhance.Contrast(image)
    image = enhancer.enhance(2)
    
    # Apply thresholding
    image = image.point(lambda x: 0 if x < 128 else 255, "1")
    
    # Optional: Resize for better OCR (e.g., 2x the original size)
    image = image.resize((image.size[0] * 2, image.size[1] * 2), Image.ANTIALIAS)
    
    return image

# Process the images
processed_image1 = preprocess_image(image_path1)
# processed_image2 = preprocess_image("IMG_6375.jpg")
display(processed_image1)
# OCR on processed images
text1 = pytesseract.image_to_string(processed_image1)
# text2 = pytesseract.image_to_string(processed_image2)

print(text1)
# print(text2)
```

```python
import cv2
import numpy as np
from pytesseract import image_to_string

def preprocess_with_opencv(image_path):
    # Read the image
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    
    # Apply adaptive thresholding
    image = cv2.adaptiveThreshold(image, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)
    
    # Denoise the image
    image = cv2.fastNlMeansDenoising(image, h=30)
    
    # Resize the image
    scale_percent = 200  # Resize by 200%
    width = int(image.shape[1] * scale_percent / 100)
    height = int(image.shape[0] * scale_percent / 100)
    dim = (width, height)
    image = cv2.resize(image, dim, interpolation=cv2.INTER_LINEAR)
    
    return image

# Preprocess images with OpenCV
processed_image1 = preprocess_with_opencv(image_path1)
# processed_image2 = preprocess_with_opencv("IMG_6375.jpg")

display(processed_image1)
# OCR on processed images
text1 = image_to_string(processed_image1, )
# text2 = image_to_string(processed_image2, config=custom_config)

print(text1)
# print(text2)
```
